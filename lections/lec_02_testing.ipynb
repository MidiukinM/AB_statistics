### 1. Formulating a hypothesis

Conducting an A/B-test we always want to answer a question: **"Has the change produced a positive (or negative) result to our members and business itself?"**

To answer this question, we select a key metric, analyze user actions and then compare the results. Given this, we can form null and alternative hypotheses in the statistics language:

$$
\begin{aligned}
& H_0: \text{there are no changes} \\
& H_A: \text{there are changes} \\
& \\
& H_0: \bar x_A = \bar x_B \\
& H_A: \bar x_A \neq \bar x_B \\
& \\
& H_0: \bar x_A - \bar x_B = 0\\
& H_A: \bar x_A - \bar x_B \neq 0
\end{aligned}
$$

### 2. Comparing the results

### Bad way

Imagine that we managed the A/B-test and got the following results: 

$\bar x_A = 25.6 \quad \bar x_B = 28 \Rightarrow \bar x_B - \bar x_A = 2.4 > 0 \Rightarrow \text{great!!!}$

Does it mean that the new feature is better? **The answer is no** and there are two reasons for this:

1. $\bar x_A$ and $\bar x_B$ are _random variables_ that are calculated from _randomly generated samples_. If we had selected people into groups a little differently, we would have received completely different numbers. So, it is necessary to correctly assess _the degree of this randomness_.
2. we don't understand _the threshold_ by which we make a decision. How much is it for us to change the metric to $+2.4$? What if it were to $+1.4$ or $+0.5$? Is it a lot? 

### Good way

Let's look at the situation from the statistical side. We have a random variable $\bar x_B - \bar x_A$. Any random variable is arranged in its own way, that is, it has some kind of distribution. Then we can:

1. determine what distribution this variable has, if there are actually no changes (if we believe in the $H_0$ hypothesis). We call it $H_0$ distribution. 
2. see where our current value of this variable ($2.4$) got into $H_0$ distribution
3. comparing with the threshold, understand how likely it is for us and make a decision about rejecting or not rejecting the $H_0$ hypothesis

- current value > threshold $\Rightarrow$ $H_0$ is rejected
- current value < threshold $\Rightarrow$ $H_0$ is not rejected

if, according to such a procedure, we rejected the $H_0$ hypothesis that there are no changes, we say that the change is **statistically significant**

```python
import numpy as np
import scipy.stats as sts
from matplotlib import pyplot as plt
import seaborn as sns

#plt.style.use('seaborn')
%matplotlib inline
 
import warnings
warnings.filterwarnings('ignore')
```

```python
norm_rv_zero = sts.norm(loc=0, scale=2)
norm_rv_alt = sts.norm(loc=5.5, scale=2.5)

n_obs = 10**6

x_zero = norm_rv_zero.rvs(n_obs)
x_alt = norm_rv_alt.rvs(n_obs)
threshold = norm_rv_zero.ppf(0.975)
```

```python
plt.figure(figsize=(10, 5))

sns.distplot(x_zero, hist=False, label='$H_0$ distribution')
sns.distplot(x_alt, hist=False, label='$H_A$ distribution')

plt.axvline(x=5, color='green', ls='-', lw=2, label='observation value')
plt.axvline(x=threshold, color='grey', ls='--', lw=2, label='threshold')

plt.legend()
plt.show()
```

### 3. Type I and II errors

If a procedure like above shows a statistically significant positive improvement in the primary decision metric, the feature is a strong candidate for a roll out to all of our members. But it's important to acknowledge that no approach to decision making can entirely eliminate uncertainty and the possibility of making mistakes.

**Mistakes that we make:**

- `Type I error (False Positive)` $-$ when the data from the test indicates a meaningful difference between the control and treatment experiences, but in truth there is no difference. 


- `Type II error (False Negative)` $-$ when the data do not indicate a meaningful difference between treatment and control, but in truth there is a difference.


We can easily get it from the picture below: 

```python
plt.figure(figsize=(10, 5))

ax = sns.distplot(x_zero, hist=False, label='$H_0$ distribution')
ax = sns.distplot(x_alt, hist=False, label='$H_A$ distribution')

l1 = ax.lines[0]
l2 = ax.lines[1]

x1, y1 = l1.get_xydata().T
x2, y2 = l2.get_xydata().T 

ax.fill_between(x1, y1, color="blue", where=(x1 >= threshold), alpha=0.3, label='type I error')
ax.fill_between(x2, y2, color="orange", where=(x2 <= threshold), alpha=0.5, label='type II error')

plt.axvline(x=6.2, color='green', ls='-', lw=2, label='observation value')
plt.axvline(x=threshold, color='grey', ls='--', lw=2, label='threshold')

plt.legend()
plt.show()
```

- `Type I error` $-$ we see that our current value > threshold. The probability of falling into this area is quite small, so we believe that our random variable is from another $H_A$ distribution (not from the $H_0$ distribution). But despite the fact that the probability is small, it still exists, so we cannot reject the hypothesis with 100% certainty.



- `Type II error` $-$ we see that there is an orange area where we don't reject our $H_0$ hypothesis, but there is a chance that it belongs to $H_A$ distribution. 


Also from the picture we can see that there is always a trade-off between type I and type II errors $-$ the more one the less the other.

$\text{Type I error} = \mathbb{P}(H_0 \text{ is rejected} | H_0 \text{ is true})$

$\text{Type II error} = \mathbb{P}(H_0 \text{ is not rejected} | H_0 \text{ is false})$

### 4. Significance level and the power of testing procedure

**Significance level**

We already know that for making a decision we need a threshold which the type I error depends on. Therefore, we will choose this threshold ourselves by fixing the acceptable type I error (False Positive rate). It is our significance level (called $\alpha$). 

- $\alpha = \mathbb{P}(H_0 \text{ is rejected} | H_0 \text{ is true})$ $-$ the acceptable False Positive rate

Then our threshold is the quantile of $1 - \alpha$ level. 

Usually, $\alpha$ is setted to no more than $5\%$.

**Power**

The power of testing procedure is the probability of rejecting the $H_0$ hypothesis when it is false. Then we can express the power formula in terms of type II error:

- $\text{power} = 1 - \mathbb{P}(H_0 \text{ is not rejected} | H_0 \text{ is false}) = 1 - \beta$ 

Usually, $\beta$ is setted to $20\%$.

### 5. p-value

p-value is just another way of comparing the observed value with the threshold. Previously, we did this by calculating the quantiles, now with using probabilities. 

- `p-value` is the probability of obtaining test results at least as extreme as the results actually observed, under the assumption that the null hypothesis is correct.

```python
plt.figure(figsize=(10, 5))

ax = sns.distplot(x_zero, hist=False, label='$H_0$ distribution')

l1 = ax.lines[0]

x1, y1 = l1.get_xydata().T

ax.fill_between(x1, y1, color="blue", where=(x1 >= threshold), alpha=0.3, label='$\\alpha$')
ax.fill_between(x1, y1, color="green", where=(x1 >= 2.5), alpha=0.3, label='p-value')

plt.axvline(x=2.5, color='green', ls='-', lw=2, label='observation value')
plt.axvline(x=threshold, color='grey', ls='--', lw=2, label='threshold')

plt.legend()
plt.show()
```

- `p-value` < $\alpha$ $\Rightarrow$ $H_0$ is rejected
- `p-value` > $\alpha$ $\Rightarrow$ $H_0$ is not rejected
